```python
from qiskit import QuantumCircuit, assemble
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import Aer, AerSimulator
from math import pi, sqrt
sim = AerSimulator()
```

```python
state1=[0,1]
```

```python
qcx = QuantumCircuit(1)
qcx.initialize(state1,0)
qcx.x(0)
qcx.draw(initial_state=True)
```

```python
qcx.save_statevector()
resultx = sim.run(qcx).result()
statex = resultx.get_statevector()
plot_bloch_multivector(statex)
```

```python
print("Statevector:", statex)
```

```python
qcy = QuantumCircuit(1)
qcy.initialize(state1,0)
qcy.y(0)
qcy.draw(initial_state=True)
```

```python
qcy.save_statevector()
resulty = sim.run(qcy).result()
statey = resulty.get_statevector()
plot_bloch_multivector(statey)
```

```python
print("Statevector:", statey)
```
